ต่อมาเราจะมาแก้ไขในส่วนของ "คณะ" เราจะทำการรับค่า name (ซึ่งคือชื่อคณะหรือคอลัมน์ที่ต้องการจัดการ)

ตรวจสอบว่า name ตรงกับคำที่หลากหลาย (รูปแบบเขียนต่างกัน) ที่แสดงถึงคณะเดียวกันหรือไม่

ถ้าใช่ จะ แทนที่ให้เป็นชื่อคณะที่มาตรฐานเดียวกัน เช่น "วิทยาศาสตร์" หรือ "ศิลปศาสตร์"

ถ้าไม่ตรงกับกลุ่มใดเลย จะคืนค่าเดิม (name) กลับไป



```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','คณะศิลปศาสตร์','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวกรรม', 'วิศวกรรมศาสตร์','คณะวิศวกรรมศาสตร์','วิศวฯ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df_master['คณะ'] = df_master['คณะ'].apply(standardize_faculty)
```

ต่อมาเราจะใช้ IQR (Interquartile Range Method) ในการตรวจจับ Outlier (ค่าผิดปกติ) ในข้อมูล โดยเฉพาะข้อมูลตัวเลข เช่น "อายุ" หรือ "รายได้"

เราจะกำหนดขอบเขตซึ่ง Q1 → ควอไทล์ที่ 1 (25% แรกของข้อมูล)

Q3 → ควอไทล์ที่ 3 (75% แรกของข้อมูล) ตัวเลข 1.5 เป็นค่ามาตรฐานที่นิยมใช้ในการหาค่าผิดปกติ

IQR = Q3 - Q1 → ช่วงระหว่างควอไทล์ (Interquartile Range) ใช้วัดว่าข้อมูลช่วงกลางกระจายมากแค่ไหน

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

lower_bound และ upper_bound คือ ขอบเขตล่างและบนสำหรับตัดสินว่าค่าใดเป็นค่าผิดปกติ

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

ต่อมาเราจะกรองเฉพาะแถวที่มีอายุอยู่ในช่วงที่กำหนด

& คือตัวดำเนินการ AND ใน pandas (ไม่ใช่ and)

วิธีนี้จะช่วยตัดค่าที่ออกนอกกลุ่มข้อมูลส่วนใหญ่มากเกินไป ซึ่งก็คือหากเจอค่า "อายุ" ที่เป็น outlier ตามเกณฑ์ IQR จึงถูกลบออกนั่นเอง

```python
df_master = df_master[(df_master['อายุ'] >= lower_bound) & (df_master['อายุ'] <= upper_bound)]
```
